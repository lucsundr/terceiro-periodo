```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Logs de latência simulados (em ms)
latencias = [120, 150, 110, 130, 450, 125, 140, 135, 120, 115]

# 2. Criando o DataFrame
df = pd.Series(latencias)

# 3. Exercício: Analise o resumo estatístico
# Note o 'max' vs a 'média'. O outlier de 450ms puxa tudo pra cima!
print(df.describe())

# 4. Visualização rápida
plt.hist(latencias, bins=5, color='royalblue', edgecolor='white')
plt.title('Distribuição de Performance')
plt.show()
```
