```python
import pandas as pd

# Tempos de resposta a um servidor com um grande outlier (10 segundos)
tempos = [0.1, 0.12, 0.11, 0.13, 0.1, 10.0]

df = pd.Series(tempos)

print(f'Média: {df.mean():.2f}')
print(f'Mediana: {df.median():.3f}')
```

```python
# Versões de navegadores em um log
navegadores = ['Chrome','Firefox','Chrome','Safari','Chrome','Edge','Firefox', 'Firefox']
df_nav = pd.Series(navegadores)

print(f"Navegador mais utilizado no escritório (moda): {df_nav.mode()[0]}")
print(f"Navegador mais utilizado no escritório (moda): {df_nav.mode()[1]}")
# Saída: Navegador mais usado (Moda): Chrome
```
