```python
import pandas as pd

# Servidor Estável vs Instável
estavel = [100, 102, 98, 101, 99]
instavel = [10, 200, 50, 150, 90]

df_e = pd.Series(estavel)
df_i = pd.Series(instavel)

print(f"Estável - Média: {df_e.mean()} | Desvio: {df_e.std():.2f}")
print(f"Instável - Média: {df_i.mean()} | Desvio: {df_i.std():.2f}")
```

```python
import pandas as pd

# Lista de temperaturas da CPU (em °C)
temperaturas = [72, 74, 71, 95, 68, 110, 73, 70, 75, 72]

serie = pd.Series(temperaturas)

media = serie.mean()
desvio_padrao = serie.std()
cv = (desvio_padrao / media) * 100

print(f"Média: {media:.2f}°C")
print(f"Desvio Padrão: {desvio_padrao:.2f}°C")

# Requisito: alertar se CV > 20%
if cv > 20:
  print("Alerta, a variação da temperatura da CPU está num nivel crítico")
```
