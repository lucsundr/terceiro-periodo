```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# 1. Gerar dados aleatórios (1000 valores com média 50 e desvio padrão 10)
np.random.seed(42)
dados = np.random.normal(50, 10, 1000)

# 2. Criar o histograma
plt.hist(dados, bins=30, color = 'skyblue', edgecolor = 'black', density=True, alpha=0.6)

#3 Gerar a alinha da Normal Teórica
# Criamos um intervalo de X que cubra os dados
x = np.linspace(min(dados), max(dados), 100)
# Calculamos a curva normal para esse intervalo
p = norm.pdf(x, 50, 10)



# 3. Personalizar o gráfico
plt.title('Distribuição de Frequeência com Curva Normaç Teórica')
plt.plot(x, p, color = 'red', linewidth = 2, label = 'Cruval Normal')
plt.legend()
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.grid(True)

# 4.
plt.show
```
